today we'll generate some basic mandelbrot-like things. 
we'll import some tools ahead of time

```python
# numerical
import numpy as np

# visualization
from matplotlib import pyplot
import matplotlib.colors as colors
```

lets set some constants too

```python
WIDTH = 1024    # width of our array
HEIGHT = 1024   # height of our array
MAX_ITERATIONS = 1000  # number of iterations to run
```

we'll plot this on on X-Y coordinate system, we can use np.meshgrid to generate that.

```python
x, y = np.meshgrid(
    np.linspace(-1.5, .5, HEIGHT),
    np.linspace(-1, 1, WIDTH),
)
```

now the fun part, generating the [mandelbrot set](https://en.wikipedia.org/wiki/Mandelbrot_set)

```python
c0 = x + 1j * y
c = c0.copy()

i = np.zeros((WIDTH, HEIGHT))

for n in range(1, MAX_ITERATIONS+1):
    c = c*c + c0
    t = abs(c) > 2
    c[t] = np.nan
    i[t] = n
```

lets visualize what we just generated

```python
_= pyplot.imshow(
    i,
    # I think a greyscale colormap with symmetric log normalization looks nice
    cmap='gray', 
    norm=colors.SymLogNorm(vmin=i.min(), vmax=i.max(), linthresh=10)
)
```

```python
_= pyplot.imshow(
    i,
    # can play with the normalization params to emphasize the boundary
    cmap='gray', 
    norm=colors.SymLogNorm(vmin=i.min(), vmax=i.max(), linthresh=50, linscale=10)
)
```
